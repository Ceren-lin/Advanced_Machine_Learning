```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import numpy as np
from dataset import CustomDataset
```

```python
root_dir = "E:\\\\\\\\sallen_key\\1"  # 
dataset = CustomDataset(root_dir=root_dir)

#  dataset  CustomDataset 
# 
features = np.array([sample[0].numpy() for sample in dataset])  # 
labels = np.array([sample[1].numpy() for sample in dataset])  # 

# 
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

#  t-SNE 
tsne = TSNE(n_components=2, random_state=42)
features_tsne = tsne.fit_transform(features_scaled)

# 
plt.figure(figsize=(8, 8))
for i in np.unique(labels):
    plt.scatter(features_tsne[labels == i, 0], features_tsne[labels == i, 1], label=f'Class {i}')
plt.legend()
plt.title('Features distribution with t-SNE')
plt.xlabel('t-SNE feature 1')
plt.ylabel('t-SNE feature 2')
plt.savefig("E:\\\\\salley-key.png")  # 


plt.show()
```
